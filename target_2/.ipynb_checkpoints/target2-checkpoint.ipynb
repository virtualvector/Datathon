```python
import pandas  as pd
import numpy as np
```

```python
df = pd.read_excel("NAS2017_Target2.xlsx")
```

```python
df.columns
```

```python
df.head(5)
```

```python
df2=df[['Class', 'Boys_Participation_No.1',u'Girls_Participation_No','Boys_Math_ %.1',u'Girls_Math_ %.1']]
```

```python
df2.head(5)
```

```python
df_class3= df2.where(df2['Class']==3)
```

```python
df_class3.dropna(inplace=True)
df_class3
```

```python
df_class5= df2.where(df2['Class']==5)
df_class5.dropna(inplace=True)
df_class5
```

```python
df_class8= df2.where(df2['Class']==8)
df_class8.dropna(inplace=True)
df_class8
```

```python
df_class3['score_diff'] = abs(df_class3['Boys_Participation_No.1']-df_class3['Girls_Participation_No'])
df_class3['perc_diff']=abs(df_class3['Boys_Math_ %.1']-df_class3['Girls_Math_ %.1'])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sns.lmplot(x="score_diff",y="perc_diff",data=df_class3)
```

```python
df_class5['score_diff'] = abs(df_class5['Boys_Participation_No.1']-df_class5['Girls_Participation_No'])
df_class5['perc_diff']=abs(df_class5['Boys_Math_ %.1']-df_class5['Girls_Math_ %.1'])
sns.lmplot(x="score_diff",y="perc_diff",data=df_class5)
```

```python
df_class8['score_diff'] = abs(df_class8['Boys_Participation_No.1']-df_class8['Girls_Participation_No'])
df_class8['perc_diff']=abs(df_class8['Boys_Math_ %.1']-df_class8['Girls_Math_ %.1'])
sns.lmplot(x="score_diff",y="perc_diff",data=df_class8)
```
