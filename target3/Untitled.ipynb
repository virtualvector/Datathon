```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_excel("hima.xlsx")
df=df.where(df['State']=="HIMACHAL PRADESH")
```

```python
df.head()
```

```python
#sns.distplot(df['No of schools'])
```

```python
sns.barplot(x=df['District'],y=df["No of schools"])
```

```python
df_top_male_pass_perc= df.sort(["Male %"])
```

```python
df_top_male_pass_perc["total_pass_perc"]=(df_top_male_pass_perc["Male %"]+df_top_male_pass_perc["Female %"])/2.0
```

```python
df_toal_pass_least_5=df_top_male_pass_perc.sort("total_pass_perc")
```

```python
df_total_pass_top_5=df_top_male_pass_perc.sort("total_pass_perc",ascending=False)
```

```python
df_toal_pass_least_5.head(5)
```

```python
sns.barplot(x="District",y="total_pass_perc",data=df_toal_pass_least_5.head(5))
```

```python
sns.barplot(x="District",y="total_pass_perc",data=df_total_pass_top_5.head(5))
```

```python
df_max_min_dist=df_total_pass_top_5.head(5)
df_max_min_dist=df_max_min_dist.append(df_toal_pass_least_5)
```

```python
sorr=df_max_min_dist.sort("total_pass_perc")
```
